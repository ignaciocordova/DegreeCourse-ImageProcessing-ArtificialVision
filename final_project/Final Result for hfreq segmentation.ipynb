```python
import impavi
import skimage
import matplotlib.pyplot as plt
```

```python
im_hfreq = impavi.hfreq_segmentation('DIC_SN_15_L10_Sum00.tif')
original = plt.imread('DIC_SN_15_L10_Sum00.tif')
im_manual = plt.imread('DIC_SN_15_L10_Sum00_manual.png')
im_topman = plt.imread('DIC_SN_15_L10_Sum00_topman.png')
im_tscratch = plt.imread('DIC_SN_15_L10_Sum00_tscratch.png')
im_multiceg_Seg = plt.imread('DIC_SN_15_L10_Sum00_multiCellSeg.png')

plt.figure(figsize=(20,20))
plt.subplot(241)
plt.imshow(original,cmap='gray')

plt.subplot(242)
plt.imshow(im_manual,cmap='gray')

plt.subplot(245)
plt.imshow(im_topman,cmap='gray')

plt.subplot(246)
plt.imshow(im_multiceg_Seg,cmap='gray')

plt.subplot(247)
plt.imshow(im_tscratch,cmap='gray')

plt.subplot(248)
plt.imshow(im_hfreq,cmap='gray')
```

```python
final = impavi.low_pass_filter(im_hfreq,0.0001,plot=True)
final_otsu = impavi.otsu_filter(final,plot=True)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(121)
plt.imshow(im_manual,cmap='gray')
plt.subplot(122)
plt.imshow(final_otsu,cmap='gray')
```

```python
ssim = skimage.metrics.structural_similarity(im_manual,im_multiceg_Seg)
print('The ssim index is',ssim)
```
